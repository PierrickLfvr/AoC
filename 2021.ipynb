```python
print("DAY 14 : Extended Polymerization")
from collections import Counter

polymer = "COPBCNPOBKCCFFBSVHKO"
#polymer = "NNCB"
rules = {}
for line in open("inputs_2021/14.txt"):
    comb, insert = line.split("\n")[0].split("->")
    rules[comb.split(" ")[0]] = insert.split(" ")[1]


def run_steps(polymer, steps):
    for i in range(steps):
        #print("steps", i)
        new_polymer = ""
        new_polymer += polymer[0]
        for comb in range(len(polymer)-1):
            duo = polymer[comb]+polymer[comb+1]
            if duo in rules:
                new_polymer+= rules[duo]+duo[1]
            else:
                new_polymer+= duo[1]
        polymer = new_polymer

    counter = Counter(polymer)
    #print(counter)
    freq = list(counter.values())
    most_frequent = np.max(freq)
    least_frequent = np.min(freq)
    #print(polymer)
    return most_frequent - least_frequent
print("Part 1: ", run_steps(polymer, 10))


def run_steps_smart(polymer, steps):
    all_duos = {}
    for comb in range(len(polymer)-1):
        duo = polymer[comb]+polymer[comb+1]
        if duo not in all_duos:
            all_duos[duo] = 1
        else:
            all_duos[duo] += 1
    for i in range(steps):
        generation = {}
        for duo in all_duos:
            if duo in rules:
                amount = all_duos[duo]
                newduos = duo[0]+rules[duo], rules[duo]+duo[1]
                for newduo in newduos:
                    if newduo in generation:
                        generation[newduo]+=amount
                    else:
                        generation[newduo]=amount
                if duo in generation:   
                    generation[duo]-=amount
                else:
                    generation[duo] = -amount
        for key in generation:
            if key not in all_duos:
                all_duos[key] = generation[key]
            else:
                all_duos[key]+=generation[key]
    return all_duos

polymer = "COPBCNPOBKCCFFBSVHKO"
#polymer = "NNCB"                
all_duos = run_steps_smart(polymer, 40)

#Smart counter based on duos only
letter_counter = {}
for duo in all_duos:
    for letter in duo:    
        if letter not in letter_counter:
            letter_counter[letter] = all_duos[duo]
        else:
            letter_counter[letter] += all_duos[duo]
for letter in letter_counter:
    letter_counter[letter] = int(np.ceil(letter_counter[letter]/2))

print("Part 2: ", np.max(list(letter_counter.values()))- np.min(list(letter_counter.values())))
```

```python
from collections import defaultdict
import heapq as heap
import numpy as np
import collections

class Graph:
    def __init__(self):
        self.g = {}
        
    def add_edge(self, u, v, weight):
        #print("adding "+str(v)+" as neighbor of "+str(u)+" with dist "+str(weight))
        if str(u) not in self.g:
            self.g[str(u)] = {}
        self.g[str(u)][str(v)] = int(weight)

    def dijkstra(self, startingNode):
        visited = set()
        parentsMap = {}
        pq = []
        nodeCosts = defaultdict(lambda: float('inf'))
        nodeCosts[startingNode] = 0
        heap.heappush(pq, (0, startingNode))

        while pq:
            # go greedily by always extending the shorter cost nodes first
            _, node = heap.heappop(pq)
            visited.add(node)

            for adjNode, weight in self.g[node].items():
                if adjNode in visited:   continue

                newCost = nodeCosts[node] + weight
                if nodeCosts[adjNode] > newCost:
                    parentsMap[adjNode] = node
                    nodeCosts[adjNode] = newCost
                    heap.heappush(pq, (newCost, adjNode))
        
        return parentsMap, nodeCosts
```
